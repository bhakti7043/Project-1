```python
# for price
ds["Price"].unique()
```

```python
ds["Price"]=ds["Price"].str.replace(",","")
```

```python
# for kms_driven
ds["kms_driven"]=ds["kms_driven"].str.replace(",","").str.replace("kms","")
```

```python
ds["kms_driven"]=pd.to_numeric(ds["kms_driven"],errors='coerce')
```

```python
# handling outliers
ds=ds[ds["kms_driven"]>=10000]
```

```python
ds=ds[ds["kms_driven"]<=200000]
```
